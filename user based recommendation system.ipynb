```python
# Set the desired number of rows
target_rows = 5000

# Read the input CSV file
df = pd.read_csv('fashion.csv')  # Replace 'input_file.csv' with your file name

# Assuming you have a DataFrame called 'df' with a column 'ProductTitle'
# Check for duplicate product titles and keep only the first occurrence
duplicate_mask = df.duplicated(subset='ProductTitle', keep='first')

# Create a new DataFrame without duplicates
df_no_duplicates = df[~duplicate_mask]

# Remove duplicates from 'df' in place
df.drop_duplicates(subset='ProductTitle', keep='first', inplace=True)

# Print or use the modified DataFrame 'df' with duplicates removed
print(df)

df_with_index = df.reset_index()
df.reset_index(inplace=True)
df.head()
```

```python

# Create user-item rating matrix
rating_matrix = data.pivot_table(index='CustomerId', columns='ProductId', values='Quantity')

# Fill missing values with 0
rating_matrix = rating_matrix.fillna(0)

# Normalize the rating matrix by subtracting the mean rating of each user
user_ratings_mean = np.mean(rating_matrix, axis=1)
normalized_rating_matrix = rating_matrix.sub(user_ratings_mean, axis=0)

# Perform SVD
U, sigma, VT = np.linalg.svd(normalized_rating_matrix)

k = 50# Number of latent factors

U_k = U[:, :k]
sigma_k = np.diag(sigma[:k])
VT_k = VT[:k, :]

# Print the dimensions of the resulting matrices
print("U matrix :", U)
print("Sigma matrix :", sigma)
print("VT matrix :", VT)
```

```python


CustomerId = input("Enter the CustomerId: ")

# Convert the input to integer
CustomerId = int(CustomerId)

# Check if the 'CustomerId' exists in the original DataFrame
if CustomerId not in rating_matrix.index:
    print(f"Customer {CustomerId} not found.")
else:
    # Get the customer index in the matrix U_k
    customer_index = rating_matrix.index.get_loc(CustomerId)

    customer_embedding = U_k[customer_index, :]

    # Calculate predicted ratings for all items
    predicted_ratings = np.dot(np.dot(U_k, sigma_k), VT_k)

    # Get top N recommendations for the user
    N = 10 # Number of recommendations
    top_N_items_indices = np.argsort(predicted_ratings[customer_index, :])[::-1][:N]

    # Map the indices back to the 'ProductId' values
    top_N_items = rating_matrix.columns[top_N_items_indices]

    recommendations = data[data['ProductId'].isin(top_N_items)]['ProductTitle'].unique()

    # Get the similar users based on the top recommendations
    similar_users = data[data['ProductId'].isin(top_N_items)]['CustomerId'].unique()

    # Print the recommendations along with the user ID who provided the recommendation
    print(f"Top {N} recommendations for user {CustomerId}:")
    for i, recommendation in enumerate(recommendations):
        print(f"Recommended by Customer {similar_users[i]}: {recommendation}")


```

```python
import matplotlib.pyplot as plt


# Group the data by 'ProductId' and calculate the total quantity for each product
product_quantity = data.groupby('ProductId')['Quantity'].sum().reset_index()

# Sort the products by quantity in descending order
product_quantity = product_quantity.sort_values(by='Quantity', ascending=False)

# Create the bar plot
plt.figure(figsize=(10, 5))
plt.bar(product_quantity['ProductId'], product_quantity['Quantity'])
plt.xlabel('Product ID')
plt.ylabel('Quantity')
plt.title('Product Quantity')
plt.xticks(rotation=90)
plt.show()
```

```python
# Group the data by 'Category' and calculate the total quantity for each category
category_quantity = data.groupby('SubCategory')['Quantity'].sum().reset_index()

# Sort the categories by quantity in descending order
category_quantity = category_quantity.sort_values(by='Quantity', ascending=False)

# Create the bar plot
plt.figure(figsize=(10, 6))
plt.bar(category_quantity['SubCategory'], category_quantity['Quantity'])
plt.xlabel('Category')
plt.ylabel('Total Quantity Purchased')
plt.title('Highly Purchased Categories')
plt.xticks(rotation=90)
plt.show()
```

```python
# Group the data by 'Category' and 'ProductTitle' and calculate the total quantity for each product in each subcategory
subcategory_quantity = data.groupby(['Category', 'ProductTitle'])['Quantity'].sum().reset_index()

# Get the index of the rows corresponding to the maximum quantity for each subcategory
max_quantity_indices = subcategory_quantity.groupby('Category')['Quantity'].idxmax()

# Select the rows with the maximum quantity for each subcategory
most_purchased_products = subcategory_quantity.loc[max_quantity_indices]

print("Most purchased product title from each subcategory:")
print(most_purchased_products)
```
